```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(15,10))
sns.boxplot(x=data['Company'])
plt.show()
```

**from the above plot we can conclude that dell company has highest number of laptops followed by lenovo and acer and so on**

```python
plt.figure(figsize=(15,10))
sns.countplot(x='Price_euros', data=data)
plt.title("price")
plt.show()
```

```python
data[data['Company'] == 'Dell'].Inches.groupby(data.Inches).count().plot(kind='pie',autopct='%1.1f%%',figsize=(10, 10))
plt.axis('equal')
plt.title("Size of Dell laptops distribution")
plt.legend(["11.6","12.5","13.3","14.0","15.0","15.6","17.3"])
plt.show()
```

**from the above plot we understood that laptops having size 15.6 is ised more**

```python
data[data['Company'] == 'HP'].Inches.groupby(data.Inches).count().plot(kind='pie',autopct='%1.1f%%',figsize=(10, 10))
plt.axis('equal')
plt.title("Size of HP laptops distribution")
plt.legend(["11.6","12.5","13.3","14.0","15.0","15.6","17.3"])
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(data['HDD'], bins=10)
plt.title("HDD Distribution among laptops")
plt.show()
```

**from the above plot we can conclude that the size of hdd generally lies between 0-50 or 450-500**

```python
plt.figure(figsize=(8,5))
sns.histplot(data['Price_euros'], bins=5)
plt.title("Price Distribution")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(data['SSD'], bins=10)
plt.title("SSD Distribution")
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.bar(data['Price_euros'], data['Ram'])
plt.xlabel("price")
plt.ylabel("Ram size")
plt.title("Ram size By Price")

plt.show()
```

**from the above plot we can say that size of Ram is highest for the expensive laptops but it doesnot increase linearly**
